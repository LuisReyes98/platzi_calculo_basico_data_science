## Dependencias

```python
from matplotlib import cm
import numpy as np
import matplotlib.pyplot as plt
```

## Grafica de 3 dimensiones

```python
def f(x, y):
  return np.sin(x) + 2*np.cos(y)

res = 100

x = np.linspace(-4, 4, num=res)

y = np.linspace(-4, 4, num=res)

x, y = np.meshgrid(x, y)

z = f(x, y)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

surf = ax.plot_surface(x, y, z, cmap=cm.spring)

fig.colorbar(surf)

plt.show()
```

## Curvas de nivel

```python
fig2, ax2 = plt.subplots()

level_map = np.linspace(np.min(z), np.max(z), num=res)

cp = ax2.contour(x, y, z, levels=level_map, cmap=cm.spring)
fig2.colorbar(cp)
```

```python
fig2, ax2 = plt.subplots()

level_map = np.linspace(np.min(z), np.max(z), num=res)

cp = ax2.contourf(x, y, z, levels=level_map, cmap=cm.spring)
fig2.colorbar(cp)
```
