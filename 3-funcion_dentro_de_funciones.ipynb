```python
import numpy as np
import matplotlib.pyplot as plt

def g(x):
  return x**2

def f(x):
  return np.sin(x)

x = np.linspace(-10, 10, num=1000)

plt.grid()
plt.plot(x,f(x))
```

```python
plt.grid()
plt.plot(x, g(x))
```

```python
f_o_g = f(g(x))

plt.grid()
plt.plot(x, f_o_g)
```

```python
g_o_f = g(f(x))

plt.grid()
plt.plot(x, g_o_f)
```
